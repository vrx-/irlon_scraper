```python
# df = pd.read_csv(io.StringIO(resp.content.decode('utf-8')), comment='%')
df = pd.read_csv(io.StringIO(resp.content.decode('utf-8')), parse_dates=['Time (est)'], skiprows=43)#delimiter=',', comment='%')
df.rename(columns={'Time (est)':'Time'}, inplace=True)
df.set_index('Time', inplace=True)
```

```python
surf = [var for var in df.columns if ('L1S' in var) and ('quality' not in var)]
surf_q = [var for var in df.columns if ('L1S' in var) and ('quality' in var)]
bott = [var for var in df.columns if ('L1B' in var) and ('quality' not in var)]
bott_q = [var for var in df.columns if ('L1B' in var) and ('quality' in var)]
```

```python
sqvars = df[surf_q]
svars = df[surf]
sqvars.columns=svars.columns
svars.where(sqvars=='good', inplace=True)
svars.columns=[ ' '.join(var.split()[1:]) for var in svars.columns]
```

```python
bqvars = df[bott_q]
bvars = df[bott]
bqvars.columns=bvars.columns
bvars.where(bqvars=='good', inplace=True)
bvars.columns=[ ' '.join(var.split()[1:]) for var in bvars.columns]
```

```python
# plot_vars = ['CDOM (QSDE)', 'Chlorophyll (ug/L)', 'Dissolved oxygen (mg/L)', 'Nitrate (uM)','Phosphate (uM)',
#              'Phycocyanin fluorescence (RFU)','Turbidity (NTU)','Water temperature (deg C)', 'pH ( )']
plot_vars = ['Instrument depth (m)', 'Turbidity (NTU)', 'CDOM (QSDE)', 'Chlorophyll (ug/L)', 'Phycocyanin fluorescence (RFU)',
             'Nitrate (uM)','Phosphate (uM)', 'Dissolved oxygen (mg/L)', 'Water temperature (deg C)']

fig, axes = plt.subplots(9,1, figsize=(10,12), sharex=True)

for ax, var in zip (axes, plot_vars):
    svars[var].plot(c='k', ax=ax, label='surface')
    bvars[var].plot(c='k', ls=':', ax=ax, label='bottom')
    ax.set_ylabel('\n('.join(var.split(' (')))
axes[0].set_ylim(2,2.5)
axes[-1].set_xlabel('')
axes[-1].xaxis.set_major_locator(DayLocator(interval=3))
axes[-1].xaxis.set_minor_locator(DayLocator(interval=1))
axes[-1].xaxis.set_major_formatter(DateFormatter('%d %b'))
# axes[-1].set_xlim(svars.index[0], svars.index[-1])
axes[-1].set_xlim(pd.to_datetime('2021-07-12'), pd.to_datetime('2021-08-20'))

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc=[.85, .9])
fig.tight_layout()
fig.savefig('/Users/hboi/Desktop/irlo.png')
```
